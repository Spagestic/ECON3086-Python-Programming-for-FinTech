# Appendix: DownSampling

To better train the model, we select the same number of positive and negative cases.

```python
%matplotlib inline
import random
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import statistics
import numpy as np
import scipy
from scipy import stats
import seaborn as sns
```

```python
train = pd.read_csv('kaggle/train.csv')
```

The number of is_booking = True is just around 8% of all sample, which are miniority

```python
train["is_booking"].value_counts(normalize=True)
```

```python
train["is_booking"].value_counts(normalize=False)
```

We are going to pick 80000 positive and 80000 negative cases for training

```python
TARGET_CNT = 80000

from sklearn.utils import resample

train_majority = train[train.is_booking==0]
train_minority = train[train.is_booking==1]

df_majority_downsampled = resample(train_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=TARGET_CNT,     # to match minority class
                                 random_state=123) # reproducible results


df_minority_downsampled = resample(train_minority, 
                                 replace=False,    # sample without replacement
                                 n_samples=TARGET_CNT,     # to match minority class
                                 random_state=123) # reproducible results

df_downsampled = pd.concat([df_majority_downsampled, df_minority_downsampled])
```

Checking output and export the file

```python
df_downsampled.is_booking.value_counts()
```

```python
df_downsampled.to_csv("train_ds.csv", index=False)
```
